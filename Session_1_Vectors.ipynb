1/ a) **A river ﬂows at 3 mph and a rower rows at 6 mph. What heading should the rower
take to go straight across a river? **

In order to solve this problem, we can first use Pythagoras or $a^2 + b^2 = c^2$. Solving Pythagoras for the figure below we arrive at $c=3\sqrt3$. We now use the rule of cosines to determine the heading.


$a^{2} =b^{2} + c^{2} - 2bc \cdot cos(A)$

$2bc \cdot cos(A) = b^{2} + c^{2} - a^{2}$

$cos(A) = \frac { b^{2} + c^{2} - a^{2}}{2bc}$

$A = \arccos{\frac { b^{2} + c^{2} - a^{2}} {2bc}}$



```julia
using PyPlot
a = 3 * sqrt(3)
X,Y,U,V = zip([0,0,3,0], [0,0,0,a], [3,0,-3,a])
quiver(X,Y,U,V,angles="xy", scale_units="xy",scale=1)
minorticks_on()
axhline()
axvline()
xlim([-1,10])
ylim([-1,10])
annotate("3 mph", xy=(1, -0.5))
annotate("6 mph", xy=(2, 2))
annotate("a", xy=(0.2, 2))
annotate("\$3\\sqrt{3}\$", xy=(-0.8, 2))
annotate("b", xy=(1, 0.3))
annotate("c", xy=(1.5, 2))
annotate("A", xy=(2.4, 0.3))

```

```julia
a = 3 * sqrt(3)
b = 3
c = 6
# rule of cosines
A = rad2deg(acos ((b^2 + c^2 - a^2)/(2*b*c))) 
```

In conclusion, the rower will have to row at 30 degrees.

3/ **Use vectors to prove that the diagonals of a parallelogram bisect each other.**

```julia
using PyPlot
X,Y,U,V = zip([0,0,3,0], [0,0,3,3], [3,0,3,3], [3,3,3,0])
quiver(X,Y,U,V,angles="xy", scale_units="xy",scale=1)
A,B,C,D  = zip([3,0,0,3], [6,3,-6,-3])
quiver(A,B,C,D,angles="xy", scale_units="xy",scale=1)
minorticks_on()
xlim([-1,7])
ylim([-1,7])
annotate("A", xy=(-0.3, 0))
annotate("B", xy=(3, -0.4))
annotate("C", xy=(6, 3))
annotate("D", xy=(3, 3.1))
annotate("M", xy=(2.7, 1.7))
annotate("M'", xy=(3.2, 1.2))
```

We must prove that $\vec{AM} = \vec{AM'}$

\begin{equation*}
\vec{AM} = \frac{1}{2}\vec{AC} \\
\vec{AM'} = \vec{AB} + \frac{1}{2}\vec{BD} \\
\vec{AM'} = \frac{1}{2}\vec{AB} + \frac{1}{2}\vec{AB} + \frac{1}{2}\vec{BD} \\
\vec{AM'} = \frac{1}{2}\vec{AB} + \frac{1}{2}\vec{DC} + \frac{1}{2}\vec{BD} \\
\vec{AM'} = \frac{1}{2}\vec{AB} + \frac{1}{2}\vec{BD} + \frac{1}{2}\vec{DC}  \\
\vec{AM'} = \frac{1}{2}\vec{AC} \\
\end{equation*}

4/ **Prove using vector methods that the midpoints of the sides of a space quadrilateral for a parallelogram.**

```julia
using PyPlot
X,Y,U,V = zip([0,0,3,0], [0,0,0,3], [3,0,0,3], [0,3,3,0])
quiver(X,Y,U,V,angles="xy", scale_units="xy",scale=1)
A,B,C,D  = zip([0,1.5,1.5,1.5], [1.5,0,1.5,1.5])
quiver(A,B,C,D,angles="xy", scale_units="xy",scale=1)
minorticks_on()
xlim([-1,4])
ylim([-1,4])
annotate("A", xy=(1.3, -0.3))
annotate("B", xy=(-0.2, 1))
annotate("C", xy=(1.3, 3.2))
annotate("D", xy=(3.2, 1.5))
```

We must prove that $\frac{1}{2}\vec{B} + \frac{1}{2}\vec{C} = \frac{1}{2}\vec{A} + \frac{1}{2}\vec{D}$

By definition, we know

$\vec{B} + \vec{C} = \vec{A} + \vec{D}$

Divide each side of the equation in half:

$\frac{1}{2}(\vec {B} + \vec {C}) = \frac{1}{2}(\vec {A} + \vec {D})$

and we are left with

$\frac{1}{2}\vec{B} + \frac{1}{2}\vec{C} = \frac{1}{2}\vec{A} + \frac{1}{2}\vec{D}$
